```python
dna = "ATGCCGT"
def gc_content(sequence):
    sequence = sequence.upper()
    if len(sequence) == 0:
        return 0
    gc = sequence.count("G") + sequence.count("C")
    return (gc / len(sequence)) * 100
def base_count(sequence):
    sequence = sequence.upper()
    return {
        "A": sequence.count("A"),
        "T": sequence.count("T"),
        "G": sequence.count("G"),
        "C": sequence.count("C")
    }
dna = "ATGCCGT"

print("GC Content:", gc_content(dna), "%")
print("Base Count:", base_count(dna))
```
